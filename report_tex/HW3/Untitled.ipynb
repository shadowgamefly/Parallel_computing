```python
import matplotlib.pyplot as plt
```

```python
time1 = [30*200/1671, 50*100/1671, 85*20/1671]
time2 = [56*200/1671/2, 97*100/1671/2, 157*20/1671/2]
time4 = [266*200/1671/4, 474*100/1671/4, 773*20/1671/4]
time1 = [1/x for x in time1]
time2 = [1/x for x in time2]
time4 = [1/x for x in time4]



ghost = [200,100,20]
```

```python
plt.plot(ghost, time1, marker='o', linestyle='--', color='r', label='1 iter')
plt.plot(ghost, time2, marker='o', linestyle='--', color='r', label='2 iter')
plt.plot(ghost, time4, marker='o', linestyle='--', color='r', label='4 iter')
```

```python
plt.xlabel('# of tasks')
plt.ylabel('efficiency')
plt.title('1 single ghost cell')
plt.legend()
plt.show()
```
